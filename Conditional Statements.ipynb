#### 1. Write a Python program to find those numbers which are divisible by 3 and are between 1000 and 2500 (both included).

```python
numbers_divisible_by_3 =  []
for number in range(1000,2501):
  if number % 3 == 0:
    numbers_divisible_by_3.append(number)
print(numbers_divisible_by_3)
```

#### 2. Write a Python program to convert month name to a number of days in this months.

```python
month = input("Write the name of the month with the first letter capitalized: ")
days_of_the_month = {
    "January":31,
    "February":28,
    "March":31,
    "April":30,
    "May":31,
    "June":30,
    "July":31,
    "August":31,
    "September":30,
    "October":31,
    "November":30,
    "December":31,
}
if month in days_of_the_month:
  print(f"{month} has {days_of_the_month[month]} days.")
else:
    print("Invalid month. Make sure to write it correctly.")
```

#### 3. Write a Python program that accepts a word from the user and reverse it.

```python
word = input("Write a word: ")
reversed_word = word[::-1]
print("Reversed word: ", reversed_word)
```

#### 4. Write a Python program which iterates the integers from 1 to 50. For multiples of three print "Wonderful" instead of the number and for the multiples of five print "Nice". For numbers which are multiples of both three and five print "Perfect".

```python
for i in range(1,51):
  if i % 3 == 0:
    print(i,"Wonderful")
  elif i % 5 == 0:
    print(i,"Nice")
  elif i % 3 == 0 and i % 5 == 0:
    print(i,"Perfect")
```
